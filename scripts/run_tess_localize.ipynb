```python
import TESS_Localize as tl
import lightkurve as lk
import astropy.units as u
import warnings
warnings.filterwarnings("ignore")


```

```python
def convert_day_to_muHz(x):
    return (1./(x*u.day.to(u.s))*u.Hz).to(u.uHz).value
```

```python
day_list_405924915 = [2.87290182*2, 2.92809810014*2 ] #405924915

frequency_list_1 = [convert_day_to_muHz(i) for i in day_list_405924915]

star = 405924915
# star = 451423147

search_result = lk.search_tesscut('TIC %s'%star)
tpf = search_result.download(quality_bitmask='default')
print(len(tpf))
result_1 = tl.Localize(targetpixelfile=tpf, gaia=True, magnitude_limit=18,
                               frequencies=frequency_list_1, frequnit=u.uHz, principal_components = 3)
```

```python
result_1.pca()
```

```python
result_1.info()
```

```python
result_1.plot_lc()
```

```python
for i in range(len(frequency_list_1)):
    result_1.plot(frequencylist_index=i,method='snr')
```
